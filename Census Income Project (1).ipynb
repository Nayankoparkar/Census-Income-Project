```python
out_cols = ['age','fnlwgt','education-num','hours-per-week'] # here i satrted removing outliers
```

```python
for col in out_cols:
  q1 = df[col].quantile(0.25)
  q3=df[col].quantile(0.75)
  iqr=q3-q1
  ul = q3+ 1.5 *iqr
  ll = q1-1.5*iqr
  df = df[(df[col] <= ul) & (df[col] >= ll)]
```
